```python
image_shape = (300, 300,3)
```

```python
train_data = metadata[metadata['Dataset_type']=='TRAIN']
test_data = metadata[metadata['Dataset_type']=='TEST']
```

```python
train_data
```

```python
train_data.shape
```

```python
train_fill = train_data.fillna('unknown')
test_fill = test_data.fillna('unknown')
```

```python
final_train_data = train_data[(train_data['Label']=='Normal')| ((train_data['Label']=='Pnemonia') & 
                                                                (train_data['Label_2_Virus_category']=='COVID-19'))]
```

```python
final_train_data['target'] = ['negative' if holder == 'Normal' else 'positive' for holder in final_train_data['Label']]
```

```python
final_train_data.head()
```

```python
final_train_data = shuffle(final_train_data, random_state=1)

final_validation_data = final_train_data.iloc[1000:, :]
final_train_data = final_train_data.iloc[:1000, :]
```

```python
train_folder = 'C:\\Users\\Aviral Pamecha\\Desktop\\Deep learning course\\Datasets\\COVID-19\\coronahack-chest-xraydataset\\Coronahack-Chest-XRay-Dataset\\Coronahack-Chest-XRay-Dataset\\train\\'
```

```python
test_folder = 'C:\\Users\\Aviral Pamecha\\Desktop\\Deep learning course\\Datasets\\COVID-19\\coronahack-chest-xraydataset\\Coronahack-Chest-XRay-Dataset\\Coronahack-Chest-XRay-Dataset\\test\\'
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_image_gen = ImageDataGenerator(
rescale=1./255,
    featurewise_center=True,
    featurewise_std_normalization=True,
    rotation_range=90,
    width_shift_range=0.15,
    height_shift_range=0.15,
    horizontal_flip=True,
    zoom_range=[0.9, 1.25],
    brightness_range=[0.5, 1.5])
```

```python

test_image_gen= ImageDataGenerator(
    rescale=1./255
)
```

```python
train_generator = train_image_gen.flow_from_dataframe(
    dataframe=final_train_data,
    directory=train_folder,
    x_col='X_ray_image_name',
    color_mode = 'rgb',
    y_col='target',
    target_size=image_shape[:2],
    batch_size=16,
    seed=2020,
    shuffle=True,
    class_mode='binary'
)
```

```python
validation_generator = train_image_gen.flow_from_dataframe(
    dataframe=final_validation_data,
    directory=train_folder,
    x_col='X_ray_image_name',
    y_col='target',
    target_size=image_shape[:2],
    batch_size=8,
    seed=2020,
    shuffle=True,
    class_mode='binary'
)
```

```python

test_generator = test_image_gen.flow_from_dataframe(
    dataframe=test_data,
    directory=test_folder,
    x_col='X_ray_image_name',
    target_size=image_shape[:2],
    shuffle=False,
    batch_size=16,
    class_mode=None
)
```

```python
from tensorflow.keras.metrics import *

METRICS = [
      TruePositives(name='tp'),
      FalsePositives(name='fp'),
      TrueNegatives(name='tn'),
      FalseNegatives(name='fn'), 
      BinaryAccuracy(name='accuracy'),
      Precision(name='precision'),
      Recall(name='recall'),
      AUC(name='auc'),
]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense, Conv2D, MaxPooling2D
```

```python
model = Sequential()
```

```python
model.add(Conv2D(filters=64, kernel_size=(3,3),input_shape=image_shape, activation='relu',))
model.add(MaxPooling2D(pool_size=(3, 3)))

model.add(Conv2D(filters=32, kernel_size=(3,3),input_shape=image_shape, activation='relu',))
model.add(MaxPooling2D(pool_size=(3, 3)))

model.add(Conv2D(filters=32, kernel_size=(3,3), input_shape=image_shape, activation='relu',))
model.add(MaxPooling2D(pool_size=(3, 3)))


model.add(Flatten())


model.add(Dense(128))
model.add(Activation('relu'))


model.add(Dropout(0.5))


model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=METRICS)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=2)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
model.summary()
```

```python
history = model.fit_generator(train_generator,epochs=20,
                              validation_data=validation_generator,
                             callbacks=[early_stop])
```
